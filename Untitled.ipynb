```python
import albumentations as A
from albumentations.pytorch import ToTensorV2
import numpy as np

# Define the image size we want to train with
IMG_SIZE = 512

# Define the strong augmentation pipeline for the training set
train_transform = A.Compose([
    # Resize the image and mask to a slightly larger size before cropping
    A.Resize(int(IMG_SIZE * 1.25), int(IMG_SIZE * 1.25)),
    # Randomly crop to the final target size
    A.RandomCrop(width=IMG_SIZE, height=IMG_SIZE),
    # Randomly flip the image and mask horizontally
    A.HorizontalFlip(p=0.5),
    # Apply color distortions to the image only
    A.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.2, p=0.5),
    # Normalize the image with ImageNet statistics
    A.Normalize(mean=(0.485, 0.456, 0.406), std=(0.229, 0.224, 0.225)),
    # Convert the image and mask to PyTorch Tensors
    ToTensorV2(),
])

# Define a simpler pipeline for the validation set (no random augmentations)
val_transform = A.Compose([
    A.Resize(width=IMG_SIZE, height=IMG_SIZE),
    A.Normalize(mean=(0.485, 0.456, 0.406), std=(0.229, 0.224, 0.225)),
    ToTensorV2(),
])

print("✅ Augmentation pipelines defined.")
```

```python
from torch.utils.data import Dataset

class FoodSegPyTorchDataset(Dataset):
    def __init__(self, hf_dataset, transform=None):
        self.hf_dataset = hf_dataset
        self.transform = transform

    def __len__(self):
        return len(self.hf_dataset)

    def __getitem__(self, idx):
        # Get one raw example (image and label as PIL images)
        example = self.hf_dataset[idx]
        image = example['image']
        label = example['label']

        # Convert PIL images to numpy arrays, which albumentations expects
        image_np = np.array(image)
        label_np = np.array(label)

        # Apply transformations
        if self.transform:
            transformed = self.transform(image=image_np, mask=label_np)
            image = transformed['image']
            # Ensure the mask is a LongTensor for the loss function
            label = transformed['mask'].long()

        return image, label

# Create the PyTorch-compatible datasets
train_pytorch_dataset = FoodSegPyTorchDataset(ds['train'], transform=train_transform)
val_pytorch_dataset = FoodSegPyTorchDataset(ds['validation'], transform=val_transform)

print("✅ PyTorch Dataset classes created.")
print(f"Number of training samples: {len(train_pytorch_dataset)}")
print(f"Number of validation samples: {len(val_pytorch_dataset)}")
```

```python
import torch
import torch.nn as nn
from torch.optim import AdamW
from torch.optim.lr_scheduler import PolynomialLR
import sys
import os
from tqdm import tqdm # Use the standard tqdm for rich progress bars
import numpy as np
import timm

# --- STEP 0: Correctly Add the Cloned Repos to Python's Path ---
# This ensures we can import from the swin-tuna and mmsegmentation repos.
# Adjust these paths to where you have cloned the repositories.
SWIN_TUNA_PATH = os.path.abspath('./swin_tuna') 
if SWIN_TUNA_PATH not in sys.path:
    sys.path.append(SWIN_TUNA_PATH)

# Now we can import correctly from the libraries
from mmseg.models.decode_heads import UPerHead # CORRECTED class name
from swin_tuna.modeling.tuna import Tuna
from swin_tuna.modeling.tuna_injector import inject_attributes 
from swin_tuna.utils.model_utils import freeze_model, c2_xavier_fill

# --- (You would have your DataLoaders and num_labels from the previous steps) ---
# --- For demonstration, we'll create dummy versions ---
num_labels = 104
IMG_SIZE = 384 # We will now correctly handle this size
# Dummy train dataloader: 10 batches of size 2
dummy_train_images = torch.randn(20, 3, IMG_SIZE, IMG_SIZE)
dummy_train_masks = torch.randint(0, num_labels, (20, IMG_SIZE, IMG_SIZE))
train_dataset = torch.utils.data.TensorDataset(dummy_train_images, dummy_train_masks)
train_dataloader = torch.utils.data.DataLoader(train_dataset, batch_size=2)
# Dummy validation dataloader: 5 batches of size 2
dummy_val_images = torch.randn(10, 3, IMG_SIZE, IMG_SIZE)
dummy_val_masks = torch.randint(0, num_labels, (10, IMG_SIZE, IMG_SIZE))
val_dataset = torch.utils.data.TensorDataset(dummy_val_images, dummy_val_masks)
val_dataloader = torch.utils.data.DataLoader(val_dataset, batch_size=2)
# -----------------------------------------------------------------------------


# --- STEP 1: Build the SOTA Model ---
print("--- Building SOTA Model ---")

# 1a. Create Backbone with CORRECT Image Size
print("Step 1: Creating the SwinV2 backbone...")
model_name = "swinv2_large_window12to24_192to384" 
backbone = timm.create_model(
    model_name,
    pretrained=True,
    features_only=True,
    out_indices=(0, 1, 2, 3),
    img_size=IMG_SIZE  # <-- THIS IS THE CRUCIAL FIX for the AssertionError
)
backbone_feature_channels = backbone.feature_info.channels()
print(f"Backbone created for input size {IMG_SIZE}x{IMG_SIZE}. Feature channels: {backbone_feature_channels}")


# 1b. Define and run our custom TUNA injector
def inject_tuna_into_timm_swinv2(target_model: nn.Module):
    conv_size_list = [5, 5, 5, 3]
    hidden_dim_list = [64, 64, 96, 192]
    # def tuna_swin_v2_block_forward(self, x):
    #     identity = x
    #     # print(x)
    #     # print(x.shape)
    #     B, H, W, C = x.shape
    #     x = x.view(B, H*W, C)
    #     x_attn = self.attn(x)
    #     x = identity + self.drop_path1(self.norm1(x_attn))
    #     x_after_attn = self.tuna_1(x)
    #     x = x + self.drop_path1(x_after_attn)
    #     identity = x
    #     x_mlp = self.mlp(x)
    #     x = identity + self.drop_path2(self.norm2(x_mlp))
    #     x_after_ffn = self.tuna_2(x)
    #     x = x + x_after_ffn
    #     return x
    def tuna_swin_v2_block_forward(self, x):
    # x is NHWC here, by Swin-V2 design
        B, H, W, C = x.shape
    
        # Attention path with proper window partitioning
        attn_out = self._attn(x)  # <-- IMPORTANT: use _attn, not self.attn
        x = x + self.drop_path1(self.norm1(attn_out))
    
        # TUNA after attention, operate on (B, N, C)
        x = x.view(B, -1, C)
        # x = x + self.drop_path1(self.tuna_1(x))
        x = x + self.drop_path1(self.tuna_1(x, (H, W)))
    
        # FFN (Swin-V2 uses post-norm on MLP output)
        ff = self.mlp(x)
        x = x + self.drop_path2(self.norm2(ff))
    
        # Optional: TUNA after FFN
        # x = x + self.tuna_2(x)
        x = x + self.tuna_2(x, (H, W))
    
        # Back to NHWC for next layers
        # x = x.view(B, H, W, C)
        x = x.view(B, H, W, C)
        # x = x.permute(0, 3, 1, 2).contiguous()
        print(x.shape)
        return x

    stages = [target_model.layers_0, target_model.layers_1, target_model.layers_2, target_model.layers_3]
    for i, stage in enumerate(stages):
        # print(x.shape)
        for block in stage.blocks:
            # print(x.shape)
            dim = block.mlp.fc1.in_features
            inject_attributes(block, conv_size=conv_size_list[i], dim=dim, hidden_dim=hidden_dim_list[i])
            block.forward = tuna_swin_v2_block_forward.__get__(block, type(block))
inject_tuna_into_timm_swinv2(backbone)
print("TUNA injection complete.")


# 1c. Create UPerNet Head and initialize with c2_xavier_fill
head = UPerHead( # CORRECTED Class Name
    in_channels=backbone_feature_channels, in_index=[0, 1, 2, 3],
    pool_scales=(1, 2, 3, 6), channels=512, dropout_ratio=0.1,
    num_classes=num_labels, norm_cfg=dict(type='BN', requires_grad=True), 
    align_corners=False,
    loss_decode=dict(type='CrossEntropyLoss', use_sigmoid=False, loss_weight=1.0)
)
print("Initializing UPerNet head with c2_xavier_fill...")
for m in head.modules():
    if isinstance(m, nn.Conv2d):
        c2_xavier_fill(m)

# 1d. Assemble final model
class SOTAModel(nn.Module):
    def __init__(self, backbone, head):
        super().__init__()
        self.backbone = backbone
        self.head = head
    def forward(self, x):
        features = self.backbone(x)
        features = [f.permute(0, 3, 1, 2).contiguous() for f in features]  # NHWC -> NCHW
        logits = self.head.forward(features)
        return logits
model = SOTAModel(backbone, head)

# 1e. Freeze parameters correctly
freeze_model(model)
for name, param in model.named_parameters():
    if 'tuna' in name or 'head' in name:
        param.requires_grad = True
print("Model assembled and parameters frozen.")


# --- STEP 2: Setup Training Components ---
print("\n--- Setting up Training Components ---")
if torch.cuda.is_available():
    device = torch.device("cuda")
    print("Training on GPU:", torch.cuda.get_device_name(0))
else:
    device = torch.device("cpu")
    print("Training on CPU")
model.to(device)

LEARNING_RATE = 0.0001
WEIGHT_DECAY = 0.05
NUM_EPOCHS = 10
TOTAL_ITERATIONS = NUM_EPOCHS * len(train_dataloader)
criterion = nn.CrossEntropyLoss()
optimizer = AdamW(model.parameters(), lr=LEARNING_RATE, betas=(0.9, 0.999), weight_decay=WEIGHT_DECAY)
scheduler = PolynomialLR(optimizer, total_iters=TOTAL_ITERATIONS, power=0.9)
print(f"Optimizer: AdamW | Scheduler: PolynomialLR")


# --- STEP 3: The Training & Validation Loop ---
print("\n--- Starting Training ---")
for epoch in range(NUM_EPOCHS):
    model.train()
    train_loss = 0.0
    
    # Use tqdm for a detailed training progress bar
    train_progress_bar = tqdm(train_dataloader, desc=f"Epoch {epoch+1}/{NUM_EPOCHS} [Train]", leave=False)
    
    for images, masks in train_progress_bar:
        images, masks = images.to(device), masks.to(device)
        
        optimizer.zero_grad()
        outputs = model(images)
        outputs = nn.functional.interpolate(outputs, size=masks.shape[-2:], mode='bilinear', align_corners=False)
        loss = criterion(outputs, masks)
        loss.backward()
        optimizer.step()
        scheduler.step()
        
        train_loss += loss.item()
        train_progress_bar.set_postfix(loss=f"{loss.item():.4f}", lr=f"{scheduler.get_last_lr()[0]:.6f}")

    avg_train_loss = train_loss / len(train_dataloader)
    print(f"Epoch {epoch+1}/{NUM_EPOCHS} -> Training Loss: {avg_train_loss:.4f}")

    # --- Validation Loop with tqdm ---
    model.eval()
    val_loss = 0.0
    
    val_progress_bar = tqdm(val_dataloader, desc=f"Epoch {epoch+1}/{NUM_EPOCHS} [Val]", leave=False)
    
    with torch.no_grad():
        for images, masks in val_progress_bar:
            images, masks = images.to(device), masks.to(device)
            outputs = model(images)
            outputs = nn.functional.interpolate(outputs, size=masks.shape[-2:], mode='bilinear', align_corners=False)
            loss = criterion(outputs, masks)
            val_loss += loss.item()
            val_progress_bar.set_postfix(loss=f"{loss.item():.4f}")
            
    avg_val_loss = val_loss / len(val_dataloader)
    print(f"Epoch {epoch+1}/{NUM_EPOCHS} -> Validation Loss: {avg_val_loss:.4f}")

print("\n--- Training Complete ---")
```
